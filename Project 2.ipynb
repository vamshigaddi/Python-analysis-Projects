```python
import pandas as pd
```

```python
import numpy as np
```

```python
import seaborn as sns
```

```python
df=pd.read_csv("C:/Users/vamsh/Downloads/Diwali Sales Data.csv",encoding="unicode_escape")
```

```python
df.drop(["Status","unnamed1"],axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df.notnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df["Amount"]=df["Amount"].astype(int)
```

```python
df["Amount"].dtype
```

```python
df.describe()
```

```python
# Expolring data analysis
```

```python
# Gender
```

```python
gx=sns.countplot(x="Gender",data=df)
for bars in gx.containers:
    gx.bar_label(bars)
```

```python
sale_gen=df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x= "Gender", y= "Amount",hue="Gender",data=sale_gen)
```

```python
#AGE GROUP
```

```python
ax=sns.countplot(x="Age Group",data=df,hue="Gender")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sale_age=df.groupby(["Age Group","Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Age Group",y="Amount",hue="Gender",data=sale_age)
```

```python
#Marital Status
```

```python
mx=sns.countplot(x="Marital_Status",hue="Gender",data=df)
for bars in mx.containers:
    mx.bar_label(bars)
```

```python
sale_m=df.groupby(["Marital_Status","Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Marital_Status",y="Amount",hue="Gender",data=sale_m)
```

```python
# Occupation
```

```python
plt.figure(figsize=(25,5))
ox=sns.countplot(x="Occupation",data=df)
for bars in ox.containers:
    ox.bar_label(bars)
```

```python
plt.figure(figsize=(20,6))
sale_oc=df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Occupation",y="Amount",data=sale_oc)
```

```python
# State Sales
```

```python
plt.figure(figsize=(26,6))
stx=sns.countplot(x="State",data=df)
for bars in stx.containers:
    stx.bar_label(bars)
```

```python
plt.figure(figsize=(26,6))
sale_st=df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="State",y="Amount",data=sale_st)
```
